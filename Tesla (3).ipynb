```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import yfinance as yf

# Download the html data using requests
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
html_data = requests.get(url).text

# Parse the html data using beautifulsoup
soup = BeautifulSoup(html_data, "html.parser")

# Extract the table using beautifulsoup
table = soup.find("tbody")
rows = table.find_all("tr")

# Create the dataframe
data = []
for row in rows:
    cells = row.find_all("td")
    date = cells[0].text
    revenue = cells[1].text
    revenue = revenue.replace("$", "").replace(",", "")
    data.append([date, int(revenue)])

gme_revenue = pd.DataFrame(data, columns=["Date", "Revenue"])

# Reset the index
gme_revenue.reset_index(inplace=True, drop=True)

# Get data for GME stock
gme_data = yf.download("GME")

def make_graph(data, revenue, title):
    fig, axs = plt.subplots(2, figsize=(15, 10))
    axs[0].plot(data.Close)
    axs[0].set_title("GME Stock Price")
    axs[0].set_xlabel("Date")
    axs[0].set_ylabel("Close Price")
    axs[0].xaxis.set_tick_params(rotation=45)
    axs[0].xaxis.set_tick_params(labelsize=8)
    axs[0].xaxis.set_major_locator(plt.MaxNLocator(6))
    axs[1].plot(revenue.Revenue)
    axs[1].set_title("GME Revenue")
    axs[1].set_xlabel("Date")
    axs[1].set_ylabel("Revenue (in million dollars)")
    axs[1].xaxis.set_tick_params(rotation=45)
    axs[1].xaxis.set_tick_params(labelsize=8)
    axs[1].xaxis.set_major_locator(plt.MaxNLocator(6))
    fig.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()

make_graph(gme_data, gme_revenue, 'GME')
```
